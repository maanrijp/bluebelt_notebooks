# grafieken

```python
import bluebelt as blue
import pandas as pd
```

```python
blue.set_style("fat")
blue.set_figsize_in_pixels(500, 350)
blue.set_language('nl')
blue.set_transparent(False)
```

```python
series = blue.dummy_series(1)
frame = blue.dummy_frame(1, 4)
```

### line

```python
series._.workdays(drop=True)._.line()
```

```python
frame._.workdays(drop=True)._.line()
```

### area

```python
series._.workdays(drop=True)._.area()
```

```python
frame.iloc[:90,:2]._.workdays(drop=True)._.area() # de eerste 90 dagen, de eerste 2 skills
```

### histogram

```python
series._.workdays(drop=True)._.hist()
```

```python
frame.iloc[:,[0,3]]._.workdays(drop=True)._.hist() # skills 0 en 3 (kolom 1 en 4)
```

### boxplot

```python
series._.workdays(drop=True)._.box()
```

```python
frame._.workdays(drop=True)._.box()
```

### waterfall

```python
s = pd.Series(data=[1, -0.05, -0.08, -0.12, -0.03, -0.06, 0], index=["bruto", "verzuim", "verlof", "training", "bijzonder verlof", "meetings", "netto"])
measure = ["relative"] * (len(s) - 1) + ["total"]
```

```python
s._.waterfall(measure=measure, horizontal=True, inverty=True, format_xticks="1.0%", title="bruto-netto")
```
