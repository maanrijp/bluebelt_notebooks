```python
import numpy as np
import pandas as pd
import bluebelt as blue
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
```

```python
frame = pd.read_pickle('../data/frame.pkl')
series = frame[frame.index.isocalendar().year == 2021]['skill A']._.workdays(drop=True)
```

```python
def plot(series, xlim=None, ylim=None, max_xticks=None, format_xticks=None, format_yticks=None, title=None, xlabel=None, ylabel=None, legend=True, path=None, **kwargs):
    fig, ax = plt.subplots()
    title = title or 'test'
    ax.plot(series)
    ax.set_title(title)
    plt.tight_layout()
```

```python
plot(series, title='a')
```
