```python
frame['npm_method'] = (frame['from']-frame['to']).abs() / (frame['from']+frame['to'])
frame['qds_method'] = (frame['to']-frame['from'])/frame['to']
```

```python
def nfm(x):
    f, a = x
    return abs(f-a)/((f+a)/2)
def qds(x):
    f, a = x
    return abs(f-a)/max(f,a)
```
