```python
import pandas as pd
import numpy as np
import bluebelt as blue
```

```python
nrows = 100
ncols = 6

np.random.seed(27)
series = pd.Series(np.random.weibull(7,nrows) * 100, index=pd.date_range(start='1-1-2020', periods=nrows), name='process A')
frame = pd.DataFrame(np.random.normal(loc=100, scale=20, size=ncols*nrows).reshape(nrows, ncols), index=pd.date_range(start='1-1-2020', periods=nrows), columns=[f'process {col}' for col in range(1, ncols+1)])
```

```python
series.blue.data.resample().subsize_count(count=(4)).plot()
```

```python
series.blue.data.resample().subsize_count(count=(3,4)).plot()
```

```python
series.blue.data.resample().sum().plot()
```

```python
series.blue.data.resample().value_range().plot()
```

```python
series.blue.data.resample().std().plot()
```

```python
series.blue.data.resample().min().plot()
```

```python
series.blue.data.resample().subsize_count(count=(3,4)).plot(ylim=(100,250))
```

```python
series.blue.data.group_index(rule='isoweek').subsize_count(count=(3,4)).plot(ylim=(5,None))
```

```python
# break the series
breaks = series.shape[0] // 5
loc = list(np.arange(0+breaks, series.shape[0], breaks))
series.iloc[loc] = np.nan

# break the frame
breaks = frame.shape[0] // 5
loc = list(np.arange(0+breaks, frame.shape[0], breaks))
for x in range(frame.shape[1]):
    frame.iloc[loc,x] = np.nan
```

```python
series.blue.data.resample().subsize_count(count=(4)).plot()
```
