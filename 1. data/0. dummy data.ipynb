# dummy data

```python
import bluebelt as blue
```

je kan dummy data genereren en hierbij een aantal parameters instellen

```python
# 1 jaar, 4 skills, afronden op 0.25 uren en 5% van de data weghalen (vervangen door np.nan)
blue.dummy_frame(1, 4, 0.25, 0.05)
```

```python
# en nu met de keywords
blue.dummy_frame(years=5, columns=8, resolution=0.125, frac=0.25)
```

en ook een enkele serie kan worden gegenereerd

```python
blue.dummy_series(years=1, resolution=1, frac=0.02, name="uren")
```
