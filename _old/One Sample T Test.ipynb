```python
import pandas as pd
import numpy as np
import bluebelt as blue
import scipy.stats as stats
```

```python
n = 100
np.random.seed(25)
series = pd.Series(np.random.weibull(7,n), index=pd.date_range(start='1-1-2020', periods=n), name='main process')
```

```python
ost = series.blue.test.one_sample_t(popmean=0.919)
```

```python
ost.plot(digits=3)
```
