```python
import pandas as pd
import numpy as np
import bluebelt as blue
```

```python
n = 365
np.random.seed(25)
series = pd.Series(np.random.weibull(7,n), index=pd.date_range(start='1-1-2020', periods=n), name='main process')
```

```python
series[:50].blue.graphs.rolling_std().plot()
```

```python
x = series.blue.patterns.polynomial(confidence=0.8, adjust=True)
```
