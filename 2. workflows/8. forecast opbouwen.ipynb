# 8. forecast opbouwen

```python
import bluebelt as blue
import pandas as pd
```

```python
blue.set_style("fat")
blue.set_figsize_in_pixels(700, 450)
blue.set_language('nl')
blue.set_transparent(False)
```

```python
series = blue.create.series(1)
```

```python
series.resample("4w").mean().values
```

```python
forecast = pd.Series(
    index = pd.DatetimeIndex(['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01',
                              '2023-05-01', '2023-06-01', '2023-07-01', '2023-08-01',
                              '2023-09-01', '2023-10-01', '2023-11-01', '2023-12-01',
                              '2024-01-01']),
    data = [400, 390, 380, 360, 350, 370, 380, 390, 400, 380, 370, 360, 390],
    name = "forecast"
)
```

```python
forecast
```

### interpolatie

```python
forecast = forecast.resample('1d').interpolate()
forecast._.line()
```

```python
series._.workdays()._.pattern.polynomial().plot(ylim=(0,None))
```

```python
series._.workdays(drop=True)._.pattern.polynomial().project(2023).set_pattern(forecast).plot(ylim=(0,None))
```
