# analyse van week- en maandpatroon

```python
import bluebelt as blue
```

```python
blue.set_style("fat")
blue.set_figsize_in_pixels(500, 350)
blue.set_language('nl')
blue.set_transparent(False)
```

```python
# 
series = blue.create.series(5)
```

```python
weekday = series._.pattern.weekday()
print(weekday)
weekday.plot()
```

```python
weekday.ratio
```

```python
# werkt eigenlijk alleen goed bij veel jaren
week = series._.pattern.week()
print(week)
week.plot()
```

```python
# met bijvoorbeeld 2 jaren krijg je best veel waarschuwingen
series[series.index.isocalendar().year.isin([2021,2022])]._.pattern.week()
```

```python
day = series._.pattern.day()
print(day)
day.plot()
```

en als bonus dan ook het jaar en de maanden

```python
series._.pattern.year().plot()
```

```python
series._.pattern.month().plot()
```

### weekpatroon op het jaarpatroon projecteren

```python
series = blue.create.series(1, frac=0)._.workdays()
```

```python
weekday = series._.pattern.weekday().ratio
weekday
```

```python
poly = series._.workdays()._.pattern.polynomial(shape=3).add_weekday()
```

```python
poly.plot()
```
