# planning effort

```python
import bluebelt as blue
```

```python
blue.set_style("fat")
blue.set_figsize_in_pixels(500, 350)
blue.set_language('nl')
blue.set_transparent(False)
```

```python
frame = blue.create.frame(1, 5)._.not_holiday()
```

De planning effort wordt weergegeven in drie variabelen:
- quantity: vergelijkt het totale volume tussen twee opeenvolgende weken
$$Q = \frac{100\%}{n} \sum^n_{t=1} \frac{|X_{t} - X_{t-1}|}{max(X_{t}, X_{t-1})}$$
- distribution: vergelijkt de verdeling over de werkdagen tussen twee opeenvolgende weken
- skills: vergelijkt de verdeling van de skills tussen twee opeenvolgende weken

Standaard worden zowel `0`-waarden als `np.nan` waarden overgeslagen. Dit betekent dat wanneer een dag als geheel `0` of `np.nan` uren heeft deze dag in de vergelijking tussen de opvolgende weken wordt genegeerd.

Deze instellingen kunnen worden aangepast door `skip_zero=False` en/of `skip_na=False` aan de functie toe te voegen.

```python
effort = frame._.planning.effort(skip_na=False)
print(effort)
effort.plot(ylim=(0,.5), title='planning effort 2022')
```

```python
ease = frame._.planning.ease()
print(ease)
ease.plot(title='planning ease 2022')
```
